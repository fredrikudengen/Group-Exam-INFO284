````python
import json
import re
from pathlib import Path

notebooks = [
    Path("task1.ipynb"),
    Path("task2.ipynb"),
]

def markdown_word_count(notebook_path):
    nb = json.loads(notebook_path.read_text(encoding="utf-8"))

    markdown_text = []
    for cell in nb["cells"]:
        if cell.get("cell_type") == "markdown":
            source = "".join(cell.get("source", []))
            markdown_text.append(source)

    text = "\n".join(markdown_text)

    # Remove fenced code blocks
    text = re.sub(r"```.*?```", " ", text, flags=re.DOTALL)

    # Remove inline code
    text = re.sub(r"`[^`]*`", " ", text)

    # Remove markdown links but keep visible text: [text](url) -> text
    text = re.sub(r"\[([^\]]+)\]\([^)]+\)", r"\1", text)

    # Count words
    words = re.findall(r"\b[\w]+(?:[-'][\w]+)*\b", text)

    return len(words)

total = 0

for notebook in notebooks:
    count = markdown_word_count(notebook)
    total += count
    print(f"{notebook}: {count} markdown words")

print(f"\nTotal markdown words: {total}")
````
